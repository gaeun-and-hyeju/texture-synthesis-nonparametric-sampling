### 여기에서 당신의 입력이 필요한 부분입니다! :)
##### 그럼, 이 모든 매개변수는 도대체 무엇인가요? 물어봐 주셔서 감사합니다!
* **exampleMapPath** - 생성하고 싶은 예제 이미지의 경로를 나타내는 문자열입니다!

* **outputPath** - 생성된 이미지(들)를 저장할 경로입니다! (알고리즘은 각 생성에 대한 매개변수를 잊지 않도록 txt 파일도 생성합니다 ^^)

* **outputSize** - 생성된 이미지의 크기입니다.

* **searchKernelSize** - 각 픽셀이 이웃 픽셀을 얼마나 '인지'하는지를 나타냅니다. 값이 클수록 더 '전반적인' 구조를 포착할 수 있지만, 속도가 느려집니다.

* **truncation** - 주어진 픽셀에 대해 예제 맵에서 샘플링한 색상 후보들 중에서 어떤 것을 선택할지 결정할 때 사용됩니다. Truncation은 너무 가능성이 낮은 샘플을 선택하지 않도록 합니다. 값은 [0,1) 범위 내에 유지해야 하며, 0은 전혀 잘라내지 않음을 의미하고, 0.9는 상위 10%의 샘플만 유지하고 그 중에서 선택함을 의미합니다.

* **attenuation** - truncation과 함께 사용됩니다! attenuation은 두 번째 단계로, 더 높은 확률의 샘플을 우선시하도록 합니다 (원하는 경우 끌 수 있습니다! 값을 1로 설정하면 끌 수 있습니다). 값은 [1, ∞) 범위 내에 유지해야 하며, 만약 실험적이고 싶다면 1 미만으로 설정하여 반대로 낮은 가능성의 샘플을 우선시하도록 할 수 있습니다! (저는 시도해본 적이 없습니다)

* **snapshots** - 각 반복마다 이미지를 저장합니다 (False로 설정하면 최종 이미지만 저장됨) - GIF를 만들고 싶다면 필요합니다 :)

이게 전부입니다! 즐거운 시간 되세요 :)


```python
#PUT YOUR PARAMETERS HERE
exampleMapPath = "imgs/1.jpg"
outputSize = [75,75]
outputPath = "out/1/"
searchKernelSize = 15

textureSynthesis(exampleMapPath, outputSize, searchKernelSize, outputPath, attenuation = 80, truncation = 0.8, snapshots = True)
```

## *Make GIF!*

* **frame_every_X_steps** - sometimes you want your GIF to not include *every* frame, this one allows you to skip X number of frames! (don't worry, it will always end up on the last frame to show the fully resolved image)
* **repeat_ending** - specify how many frames the GIF will loop over the final resolved image

```python
gifOutputPath = "out/outGif_2.gif"

makeGif(outputPath, gifOutputPath, frame_every_X_steps = 15, repeat_ending = 15)
```
